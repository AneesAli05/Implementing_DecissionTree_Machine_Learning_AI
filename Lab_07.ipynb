```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('titanic_cleaned - titanic_cleaned.csv')
data.shape
```

```python
data.head()
```

```python
x=data.drop(['Survived'], axis=1)
y=data['Survived']
x.shape, y.shape
```

```python
x.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x_scaled, columns=x.columns)
x.head()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=56, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
```

```python
#creating instance of knn
clf=KNN(n_neighbors=10)
#fitting the model

clf.fit(train_x, train_y)
test_predict= clf.predict(test_x)
k=f1_score(test_predict, test_y)
print('test f1 score',k)
```

```python
def Elbow():
    test_error=[]
    for i in k:
        clf=KNN(n_neighbors=i)
        clf.fit(train_x, train_y)
        y_pred=clf.predict(test_x)
        f1= f1_score(y_pred, test_y)
        error=1-f1
        test_error.append(error)
    return test_error
```

```python
k=range(6,20, 2)
test= Elbow(k)
```

```python
#for ploting elbow function
plt.plot(k, test)
plt.xlabel('K Nieghbors')
plt.ylabel('Test error')
plt.title('Elbow curve for test')
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import mean_squared_error as mse
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=data=pd.read_csv('Bigmart_cleaned - Bigmart_cleaned.csv')
data.shape
```

```python
data.head()
```

```python
x=data.drop(['Item_Outlet_Sales'], axis=1)
y=data['Item_Outlet_Sales']
x.shape, y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x_scaled)
x.head()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=56)
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import f1_score
```

```python
#creating instance of knn
clf=KNN(n_neighbors=10)
#fitting the model
clf.fit(train_x, train_y)
test_predict= clf.predict(test_x)
k=f1_score(test_predict, test_y)
print('test f1 score',k)
```

```python
def Elbow(k):
    test_error=[]
    for i in k:
        clf=KNN(n_neighbors=i)
        clf.fit(train_x, train_y)
        y_pred=clf.predict(test_x)
        f1= f1_score(y_pred, test_y)
        error=1-f1
        test_error.append(error)
    return test_error
```

```python
k=range(6,20, 2)
test= Elbow(k)

plt.plot(k, test)
plt.xlabel('K Nieghbors')
plt.ylabel('Test error')
plt.title('Elbow curve for test')
```
